```python
x= df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Area Population','Avg. Area Number of Bedrooms']]
y=df['Price']
```

*SPLITTING TEST DATA*

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.25, random_state=10)
```

```python
xtrain=pd.DataFrame(xtrain)
xtest=pd.DataFrame(xtest)
ytrain=pd.DataFrame(ytrain)
ytest=pd.DataFrame(ytest)
```

*creating machine learning model*

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

*Traininng the model with training data *

```python

regressor=LinearRegression()
regressor.fit(xtrain,ytrain)
```

*Evaluation*

```python
regressor.score(xtest, ytest)
```

```python
print(regressor.score(xtest,ytest)*100,'%')
```

```python
print('%.1f%%'% (regressor.score(xtest,ytest)*100))
```
